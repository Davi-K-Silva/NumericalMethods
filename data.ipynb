```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cell = pd.read_csv('data.csv')
cell
```

```python
fig = plt.figure(figsize = (10,8))

y = cell["TAVG"].to_list()
x = np.arange(0, 30, 1)
print(x)
plt.plot(x, y, 'b', x, y, 'ro')
plt.title('Data')
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
